```python
from dtumathtools import *
from sympy import *
init_printing()

# Define variables
x1, x2, x3, x4 = symbols('x1 x2 x3 x4')
x, y, z, t = symbols('x y z t')
```

```python
# Function: f(x1, x2) = sqrt(|x1 * x2|)
f1 = sqrt(abs(x1 * x2))

# Visualization: Surface Plot
dtuplot.plot3d(f1, (x1, -2, 2), (x2, -2, 2), title="Scalar Function 2D")
```

```python
# Function: f = sqrt(|x1*x2|) + i*cos(x1+x2)
f3 = sqrt(abs(x1 * x2)) + I * cos(x1 + x2)

# Visualization: Real Part
print("Plotting Real Part...")
p_re = dtuplot.plot3d(re(f3), (x1, -2, 2), (x2, -2, 2), title="Real Part", show=False)

# Visualization: Imaginary Part
print("Plotting Imaginary Part...")
p_im = dtuplot.plot3d(im(f3), (x1, -2, 2), (x2, -2, 2), title="Imaginary Part", show=False)

# Show both
(p_re + p_im).show()
```

```python
# Function: f(x1, x2) = (-x2/3, x1/3)
# Mapping x1 -> x, x2 -> y for plotting clarity, or using x1/x2 directly.
f4 = Matrix([-x2/3, x1/3])

# Visualization: Quiver Plot
# Note: Using quiver_kw to style the arrows as requested in the hint
dtuplot.plot_vector(f4, (x1, -5, 5), (x2, -5, 5), 
                    title="Vector Field 2D", 
                    quiver_kw={"color": "black", "scale": 20})
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import *
from sympy.plotting import plot3d, plot3d_parametric_line

init_printing()
x1, x2 = symbols('x1 x2')
t = symbols('t')

# 1. Scalar Function
print("Plotting Scalar Function...")
plot3d(sqrt(abs(x1*x2)), (x1, -2, 2), (x2, -2, 2), title="Scalar Function")

# 2. Vector Field (Standard Matplotlib)
print("Plotting Vector Field...")
x, y = np.meshgrid(np.linspace(-5, 5, 20), np.linspace(-5, 5, 20))
u = -y/3
v = x/3
plt.figure()
plt.quiver(x, y, u, v)
plt.title("Vector Field 2D")
plt.show()
```

```python
# Function: f(x,y,z) = (x^3 + yz^2, y^3 - xz^2, z^3)
f5 = Matrix([x**3 + y*z**2, y**3 - x*z**2, z**3])

# Visualization: 3D Quiver Plot
# Hint applied: Scaling by 0.1 to prevent arrows from overlapping
dtuplot.plot_vector(0.1 * f5, (x, -2, 2), (y, -2, 2), (z, -2, 2), 
                    title="Vector Field 3D")
```

```python
# Function: r(t) = (cos(t), sin(t), t)
r = Matrix([cos(t), sin(t), t])

# Visualization: Parametric Line
dtuplot.plot3d_parametric_line(r, (t, 0, 10), title="Parametric Curve")
```
