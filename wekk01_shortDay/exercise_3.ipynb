Question A

```python
import sympy as sp

# Define variables
x1, x2 = sp.symbols('x1 x2')
x = sp.Matrix([x1, x2])

# Define constant
c = -60

# Define matrices and vectors
A1 = sp.Matrix([[11, -12], [-12, 4]])
b1 = sp.Matrix([-20, 40])

A2 = sp.Matrix([[11, 0], [-24, 4]])
b2 = b1 # b2 is same as b1

A3 = sp.Matrix([[sp.Rational(73,5), sp.Rational(-36,5)], [sp.Rational(-36,5), sp.Rational(52,5)]])
b3 = sp.Matrix([-44, 8])

# Define quadratic form function
def compute_q(A, b, c, x):
    # Formula: x.T * A * x + b.T * x + c
    quad_term = (x.T * A * x)[0]
    lin_term = (b.T * x)[0]
    return sp.expand(quad_term + lin_term + c)

# Compute expansions
q1 = compute_q(A1, b1, c, x)
q2 = compute_q(A2, b2, c, x)
q3 = compute_q(A3, b3, c, x)

# Output results
print(f"q1(x) = {q1}")
print(f"q2(x) = {q2}")
print(f"q3(x) = {q3}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot_quadratic_forms():
    # 1. Define Grid
    x = np.linspace(-10, 10, 100)
    y = np.linspace(-10, 10, 100)
    X, Y = np.meshgrid(x, y)

    # 2. Define Functions
    # q1(x) = 11x^2 - 24xy + 4y^2 - 20x + 40y - 60
    Z1 = 11*X**2 - 24*X*Y + 4*Y**2 - 20*X + 40*Y - 60
    
    # q3(x) = 14.6x^2 - 14.4xy + 10.4y^2 - 44x + 8y - 60
    Z3 = 14.6*X**2 - 14.4*X*Y + 10.4*Y**2 - 44*X + 8*Y - 60

    # 3. Plot Setup
    fig = plt.figure(figsize=(12, 10))

    # q1 Surface Plot
    ax1 = fig.add_subplot(2, 2, 1, projection='3d')
    ax1.plot_surface(X, Y, Z1, cmap='coolwarm', alpha=0.8)
    ax1.set_title('q1 Surface: Hyperbolic Paraboloid')

    # q1 Contour Plot
    ax2 = fig.add_subplot(2, 2, 2)
    CS1 = ax2.contour(X, Y, Z1, levels=20, cmap='coolwarm')
    ax2.clabel(CS1, inline=True, fontsize=8)
    ax2.set_title('q1 Level Curves: Hyperbolas')
    ax2.grid(True)
    ax2.set_aspect('equal')

    # q3 Surface Plot
    ax3 = fig.add_subplot(2, 2, 3, projection='3d')
    ax3.plot_surface(X, Y, Z3, cmap='viridis', alpha=0.8)
    ax3.set_title('q3 Surface: Elliptic Paraboloid')

    # q3 Contour Plot
    ax4 = fig.add_subplot(2, 2, 4)
    CS3 = ax4.contour(X, Y, Z3, levels=20, cmap='viridis')
    ax4.clabel(CS3, inline=True, fontsize=8)
    ax4.set_title('q3 Level Curves: Ellipses')
    ax4.grid(True)
    ax4.set_aspect('equal')

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    plot_quadratic_forms()
```
